### Dictionary

- Stores as key value pairs


- Key must be immutable
- Value can be anything

```python
dict1 = {'abacus' : 'a calculator', 'bachelor' : 'unmarried person', 'cable' : 'strong rope' }

dict1['abacus']
```

```python
dict2 = {101 : 'John', 102 : 'Smith', 103 : 'Mark', 104 : 'David'}
dict2[101]
```

```python
# You can change the values in the dictionary

dict2[101] = 'Wayne'
dict2
```

```python
# You can add values to dictionary

dict2[105] = 'Ajay'
dict2
```

```python
# You can delete items in dictionary

del dict2[105]
dict2
```

```python
# Iterating in Dictionaries

# This prints out all the keys
for i in dict2:
    print(i)
    
# This prints out all the values
for i in dict2:
    print(dict2[i])
```

```python
# Create empty dictionary

dict1 = dict()
# dict1 = { } also creates an empty dictionary
type(dict1)
dict1['Apple'] = 'red'
dict1
```

```python
# Use for loop to create dictionary

for x in range(1,6):
    dict1[x]=x**2
    
print(dict1)
```

```python
# Creating a dictionary of preset key value pairs

dict2 = dict(((1,2),(2,4),(3,6)))
# Can use [ ] too for the internal key values

dict2
```

```python
dict2 = dict(('ab','cd','ef')) # Strings are also considered lists/ iterables
dict2
```

### Zips


```python
# Zip will match the first lists items to the second

l1 = ['A','B','C']
l2 = ['apple','ball','cat']

dict3 = dict(zip(l1,l2))
dict3
```

### Enumerate

The keys are now enumerated i.e. 0,1,2,3,4,5,6,

```python
l1 = ['a','b','c']

for item in enumerate(l1):
    print(item)
```

```python
# Create a dictionary with enumerate

dict4 = dict(enumerate(l1))
dict4
```

```python
# Dictionary Comprehension

dict6 = { x: x**2 for x in range(1,5)}
dict6
```

If you want to encapsulate in tuples ie <br>
dict6 = { (x, x**2) for x in range(1,5)}
<br>
You will need to use the dict() function or else you will get a set of lists


```python
dict7 = dict((x, x**2) for x in range(1,5))
dict7
```

```python
# Zipping Comprehension

l1 = [1,2,3]
l2 = ['a','b','c']

for x,y in zip(l1,l2):
    print(x,y)
```

```python
# Doing the above in dictionary comprehension

dict8 = dict((x,y) for x,y in zip(l1,l2))
dict8
```

```python
# Can also the the above in key value pair instead of tuple, dont use dict() function

dict8 = { x:y for x,y in zip(l1,l2)}
dict8
```

```python
dict9 = { x:y for x,y in enumerate(l2)}
dict9
```

### Dictionary Challenge

Find the longest and shortest meaning in dictionary

```python
dictionary = { 'pice' : "portion of an object or of material produced by cutting", 'patch' : 'a piece of cloth or other material used to mend or strengthen a torn or weak point', 'with' : 'having or possessing', 'small' : 'less than normal'}
dictionary
```

```python
longest = ''
longestword = 0

for i in dictionary:
    if len(dictionary[i])> longestword:
        longest = i
        longestword = len(dictionary[i])
        
print(longest)
print(longestword)
```

```python
shortest = ''
shortestword = 10000000000000000000

for i in dictionary:
    if len(dictionary[i]) < shortestword:
        shortest = i
        shortestword = len(dictionary[i])
        
print(shortest)
print(shortestword)
```

### Dictionary Challenge

Convert roman numbers to numbers

```python
roman = { 'I': 1, 'V' : 5, 'X' : 10, 'L' : 50, 'C' : 100, 'D' : 500, 'M' : 1000}
roman
```

```python
number = input('Enter your Roman Number')
value = 0

i = 0

while i < len(number):
    if i < len(number) and roman[number[i+1]] > roman[number[i]]:
        value += roman[number[i+1]] - roman[number[i]]
        i += 2
    else:
        value += roman[number[i]]
        i += 1
                                                
print(value)
    
        
```
