### Dictionary

- Stores as key value pairs


- Key must be immutable
- Value can be anything

```python
dict1 = {'abacus' : 'a calculator', 'bachelor' : 'unmarried person', 'cable' : 'strong rope' }

dict1['abacus']
```

```python
dict2 = {101 : 'John', 102 : 'Smith', 103 : 'Mark', 104 : 'David'}
dict2[101]
```

```python
# You can change the values in the dictionary

dict2[101] = 'Wayne'
dict2
```

```python
# You can add values to dictionary

dict2[105] = 'Ajay'
dict2
```

```python
# You can delete items in dictionary

del dict2[105]
dict2
```

```python
# Iterating in Dictionaries

# This prints out all the keys
for i in dict2:
    print(i)
    
# This prints out all the values
for i in dict2:
    print(dict2[i])
```

```python
# Create empty dictionary

dict1 = dict()
# dict1 = { } also creates an empty dictionary
type(dict1)
dict1['Apple'] = 'red'
dict1
```

```python
# Use for loop to create dictionary

for x in range(1,6):
    dict1[x]=x**2
    
print(dict1)
```

```python
# Creating a dictionary of preset key value pairs

dict2 = dict(((1,2),(2,4),(3,6)))
# Can use [ ] too for the internal key values

dict2
```

```python
dict2 = dict(('ab','cd','ef')) # Strings are also considered lists/ iterables
dict2
```

### Zips


```python
# Zip will match the first lists items to the second

l1 = ['A','B','C']
l2 = ['apple','ball','cat']

dict3 = dict(zip(l1,l2))
dict3
```

### Enumerate

The keys are now enumerated i.e. 0,1,2,3,4,5,6,

```python
l1 = ['a','b','c']

for item in enumerate(l1):
    print(item)
```

```python
# Create a dictionary with enumerate

dict4 = dict(enumerate(l1))
dict4
```

```python
# Dictionary Comprehension

dict6 = { x: x**2 for x in range(1,5)}
dict6
```

If you want to encapsulate in tuples ie <br>
dict6 = { (x, x**2) for x in range(1,5)}
<br>
You will need to use the dict() function or else you will get a set of lists


```python
dict7 = dict((x, x**2) for x in range(1,5))
dict7
```

```python
# Zipping Comprehension

l1 = [1,2,3]
l2 = ['a','b','c']

for x,y in zip(l1,l2):
    print(x,y)
```

```python
# Doing the above in dictionary comprehension

dict8 = dict((x,y) for x,y in zip(l1,l2))
dict8
```

```python
# Can also the the above in key value pair instead of tuple, dont use dict() function

dict8 = { x:y for x,y in zip(l1,l2)}
dict8
```

```python
dict9 = { x:y for x,y in enumerate(l2)}
dict9
```

### Dictionary Looping

```python
dict1 = {'abacus' : 'a calculator', 'bachelor' : 'unmarried person', 'cable' : 'strong rope' }

# This just prints the keys
for x in dict1:
    print(x)
```

```python
for x in dict1:
    print(x, dict1[x])
    
# Instead of using dict1[x] you can also use dict1.get(x)
# Difference is that dict1[x] will throw an error if it cannot find the index while get(x) will not
# If you want the get method to show a message if key cannot be found use
# dict1.get(106, 'Invalid Key')
```

```python
# Get keys

dict1.keys()
```

```python
# Get values

dict1.values()
```

```python
# Get tuple of keys and values

dict1.items()
```

```python
for x in dict1.keys():
    print(x, dict1[x])
```

```python
for x in dict1.values():
    print(x)
```

```python
for x,y in dict1.items():
    print(x,y)
```

```python
dict1 = { 101 : 'Production', 102 : 'Accounts', 103 : 'Sales & Marketing', 104 : 'Inventory'}
```

```python
# copy() creates a shallow copy

dict2 = dict1.copy()
dict2
```

```python
dict2[102] = 'Designing'
dict2
```

```python
# Add more dictionary items with update()

dict1 = { 101 : 'Production', 102 : 'Accounts', 103 : 'Sales & Marketing', 104 : 'Inventory'}
dict2 = {105: 'Designing', 106 : 'Packaging'}

dict1.update(dict2)
dict1
```

```python
# Setdefault() method,if key is not there, set value to the default you specify
# If it key exist, it will return the value of the key and ignore your default

dict1.setdefault(101, 'Advertisement')
dict1
```

```python
dict1.setdefault(110, 'Advertisement')
dict1
```

```python
# fromkeys() all the keys that you specify will have a singular value that you have to specify

l1 = [11,22,33,44]
dict3 = {}
dict3.fromkeys(l1, "Hello World")
```

### Deleting Dictionary Values



```python
dict1 = { 101 : 'Production', 102 : 'Accounts', 103 : 'Sales & Marketing', 104 : 'Inventory'}
```

```python
dict1.pop(101)
dict1
```

```python
# If key is not present when you pop it will throw an error

dict1.pop(110)
```

```python
# Leave a default message there in case index is not there

dict1.pop(110, "Index was not in Dictionary")
```

```python
dict1
```

```python
# popitem() will remove the last key value pair in the dictionary
dict1.popitem()
dict1
```

```python
# Clear() removes all items in the dictionary

dict1.clear()
dict1
```
