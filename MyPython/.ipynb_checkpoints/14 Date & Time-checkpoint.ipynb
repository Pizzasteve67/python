### Date & Time

Modules exist for 
- datetime
- time
- calendar


```python
import time
import datetime

print(time.time()) # epoch time since 1 jan 1970
print(time.localtime()) # year, month, day, hour, min, sec, no. of day (Mon:1), day of year, daylightsavings

print(time.ctime()) # Returns in string
print(datetime.datetime.now()) 
```

### Creating a Date & Time Object

Use datetime.datetime(year,month,day,hrs,min,sec) <br>
dt1 = datetime.datetime(year,month,day,hrs,min,sec) <br>
You can use keywords also <br>
dt2 = datetime.datetime(day = 1, month = 1, year = 2011, hour = 10, minute = 10)

```python
import datetime

dt1 = datetime.datetime(2010,1,1,10,10,10)
dt2 = datetime.datetime(day = 1, month = 1, year = 2011, hour = 10, minute = 10)

print(dt2 - dt1) # Compare the time difference between 2 dates
print(dt2 > dt1) # Bool 
print(dt1 > dt2)
```

```python
# You can create just a date object

date1 = datetime.date(2015,1,1)
time1 = datetime.time(10,10,10)

dt3 = datetime.datetime.combine(date1,time1) # Combine the date and time

date2 = datetime.date(2016,1,1)
time2 = datetime.time(11,11,11)

print(date2-date1)
print(date2>date1)

# You cannot subtract time
try:
    print(time2 - time1)
except:
    print("Cannot subtract time")
    
print(time2 > time1)
```

### Date time has 2 methods to convert
.strftime() converts datetime, date, time into string <br>
.strptime() converst string into date time object <br><br>
Control characters for string to convert to object


```python
from datetime import *

dt1 = datetime(2010,1,1,10,20,30)
print(dt1)
dt1.strftime('%d %B, %A %Y.%H:%M:%S') # Use the control characters
```

```python
str1 = '10 Jan 2015, 10:15:30'
dt2 = datetime.strptime(str1,'%d %b %Y, %H:%M:%S')
print(dt2)
```

```python
date1 = datetime(2010,10,1,10,20,30)
date2 = datetime(2011,10,1,10,20,30)

td = date1 - date2 # td is of class timedelta
print(td)
divmod(td.days,30) # timedelta of days divide by 30
```

### Calendar Module



```python
from calendar import *
 
print(firstweekday()) # Default is monday
print(isleap(2023))
print(leapdays(2000,2022)) # How many leap years have been within the interval
print(weekday(2000,1,1)) # What day was this date
print(prmonth(2022,3)) # Print the month calendar
```

### Calendar Class


```python
cal = Calendar()
for i in cal.iterweekdays():
    print(i)
```

```python
for i in cal.itermonthdates(2022,2):
    print(i)
```
