### Default Arguments

You can create functions with default params. When you call the function, no need to provide the argument in the param

Defaults need to specified on the right side


```python
# Defaults are only created once and the function will not keep creating new objects

def additem( item, l = []):
    l.append(item)
    return l

# After everytime you call the function, the function will use the same object l that you created
# It doesnt make new l objects for everytime you call the object
print(additem(22))
print(additem(33))
print(additem(44))
print(additem(55))
```

### Generators
Generators are functions that help to iterate, they use the yield instead of return to iterate one step of a loop. Instead of running the while loop infinitely, this yield will save the state in the loop and iterate slowly

```python
def Days():
    l = ['mon', 'tues','wed','thu', 'fri','sat','sun']
    i = 0
    
    while True:
        x = l[i]
        i = (i + 1) % 7
        yield x
        
d = Days()
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
```

### Local vs Global Variables

- Global variables can be accessed from anywhere in the python file
- Local variable can only be accessed from within the function

```python
# Interactions of local and global variables

g = 10

def func():
    g = 20
    print(g)  # This will call the local variable instead of the global, global is unchanged
    
    
func()
print(g)
```

```python
# A function cannot modify a global variable normally, it can only access it
# In order for your function to modify the global variable, use the global keyword

g = 10

def func():
    global g # This will allow your function to modify the global variable
    g = 20 
    print(g)  
    
    
func()
print(g)
```

```python
# Printing Local and Global Variables using their respective functions. Will return in dictionary form

a,b,c,d = 11,22,33,44

def fun1(a=15,b=25):
    x,y,z = 1,2,3
    print(locals())
    
fun1()
```

### A function that calls itself is a recursive function

e.g. factorials

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
    

fact(3)
```

```python
def f():
    global a
    print(a)
    a = "hello"
    print(a)
a = "world"
f()
print(a)
```

### Student Challenge Sum of scores ending with zero
From a list, find the sum of scores ending with zero

```python
l = [200, 456, 302, 100, 234, 678]


def sumzero(l):
    summation = 0
    for i in l:
        if i%10 == 0:
            summation += i
    return summation

sumzero(l)
            
    
```

### Student Challenge Invert a dictionary
Change keys to values and values to keys

```python
{'a':10,'b':20,'c':30,'d':40}

def invert(d):
    e = {}
    for i in d:
        e.update({d[i]:i})
    return e

invert({'a':10,'b':20,'c':30,'d':40})
```

### Student Challenge Pangram

Create a function that returns true if every single letter of the alphabet is in the string

```python
def pangram(string):
    alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    for i in alphabet:
        if i not in string.lower():
            return False
            break
    return True
    
    
pangram("The quick brown fox jumps over the lazy dog")
```

```python
# Alt solution, not as elegant but it is the one used in lesson

def pangram(string):
    alphabet = {'a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z'}
    setty = set()
    for i in string.lower():
        setty.add(i)
    if len(setty)-1 == len(alphabet):    #Remove the space bar
        return True
    else:
        return False
        
pangram("The quick brown fox jumps over the lazy dog")
        
        
```

### Student Challenge count upper and lower case letters

Create a function that takes in a string as input and counts the number of upper and lower case letters there are in the string

```python
def count(string):
    upper = 0
    lower = 0
    for i in string:
        if i.isupper() == True:
            upper += 1
        else:
            lower +=1
    print("Number of Upper is:", upper, "Number of lower is:", lower)

count("AbCd")
    
```

### Student Challenge Minimum
Create a function that takes in a variable number of posiitonal arguments and return the minimum number

```python
def minimum(*numbers, low_limit=None):
    if low_limit is None:
        print(min(numbers))
    else:
        l = []
        for i in numbers:
            if i > low_limit:
                l.append(i)
        print(min(l))
        
minimum(10,2,3,4,-5, low_limit=0)
```

### Student Challenge Flatten Nested Sequence

Flatten a nested list into a single list, use the generator function

```python
#Recursively flatten a nested list

items = [1,2,[3,4,[5,6],7],8]

def flatten(l):
    for i in l:
        if type(i) == list:
            yield from flatten(i)
        else:
            yield i
        
f = flatten(items)
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
```

### Function Returning a Function

### Closure Function



```python
def closure():
    msg = 'Hello'
    def display():        # Nested Function
        print('*' * 10)
        print(msg)        # Accessing Non local variable
        print('*' * 10)
        
    return display        # Return Nested Function

d = closure()
d()
```

```python
def closure(msg):

    def display():        # Nested Function
        print('*' * 10)
        print(msg)        # Accessing Non local variable
        print('*' * 10)
        
    return display        # Return Nested Function

d = closure('Hello World')
d()
```

### You can write a caller class as a closure function

```python
def Depts():
    depts = {'hr': 'Human Resource Department',
            'acc': 'Accounts and Finance Department',
            'sd': 'Sales and Marketing Department',
            'mkt': 'Marketing Department'}
    def dname(dept):
        return depts[dept]
    
    return dname

d = Depts()
s = d('hr')
print(s)
print(d('acc'))
```

### Decorator Function


```python
def decorator(fun):
    def wrapper(msg):
        print('a')
        fun(msg)
        print('b')
    
    return wrapper
        
def display(msg):
    print(msg)
    
d = decorator(display)
d('Hello')
```

### Lambda Functions


```python
# How we write regular functions these days look like this

def miles2km(miles):
    kms = miles * 1.6
    return kms

print(miles2km(10))
```

```python
# Instead of writing the function like that, I can write it in a lambda function

k = lambda miles: 1.6*miles
print(k(10))

# For lambda functions, you store the lambda function as a variable 
# Then you feed params into that variable

# lambda miles: replaces miles2km(miles):

```

```python
# Another example of adding 2 numbers 

def add(a,b):
    c = a + b
    return c

c = lambda a,b : a + b  # Rewriting the above as lambda function
c(10,5)
```

```python
f = lambda a,b : a if a>b else b  # return a if a > b, else return b

f(15,10)
```
