### Modes of Opening a file



- For 'r' if file doesnt exist, then it will return filenotfounderror
- For 'w' if file doesnt exist, it will create a new file
- For 'a' same as write, but file must exist, it will also add on to the existing content of the file
- For 'r+' read and write, behaves like and append
- For 'w+' write and read, will create file if file doesnt exist
- For 'x' same as write but file must not be already existing
