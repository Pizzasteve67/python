### Choice
When given a list, it will choose a random one in the lists

```python
l = [1,3,5,7,9,11,13]
print(choice(l))
print(choices(l,k=2)) # Choose 2 elements
# sample() works the same way as choices

print(sample(l,k=2))
```

```python
shuffle(l)
print(l) # Shuffled the elements in the list
```

### Get State 
Save the state of a random number and call that state again

```python
st = getstate()
print(random())
print(random())
setstate(st)
print(random()) # Returns the same value as the first state
print(random())
```

### Statistics


```python
from statistics import *
data = [1,2,3,4,5]

print(mean(data))
```

### Harmonic Mean


```python
print(harmonic_mean(data))
```

```python
print(median(data)) # If even it will take the average of the 2 middle numbers
```

```python
data = [1,2,3,4,5,6]
print(median_high(data)) # If even it will take the higher number
```

```python
print(median_low(data)) # If even it will take the lower number
```

```python
print(mode(data)) # Return most common element, 
                  #if element all appear same number of time, it will take the first
```

```python
# Standard deviation and variance
# p in front says it will take all elements in the iterable
# p stands for population

print(pstdev(data))
print(pvariance(data))
pstdev(data)**2
```

```python
# Standard deviation and variance without the p will sample data from the iterable

print(stdev(data))
print(variance(data))
pstdev(data)**2
```
