### Default Arguments

You can create functions with default params. When you call the function, no need to provide the argument in the param

Defaults need to specified on the right side


```python
# Defaults are only created once and the function will not keep creating new objects

def additem( item, l = []):
    l.append(item)
    return l

# After everytime you call the function, the function will use the same object l that you created
# It doesnt make new l objects for everytime you call the object
print(additem(22))
print(additem(33))
print(additem(44))
print(additem(55))
```

### Generators
Generators are functions that help to iterate, they use the yield instead of return to iterate one step of a loop. Instead of running the while loop infinitely, this yield will save the state in the loop and iterate slowly

```python
def Days():
    l = ['mon', 'tues','wed','thu', 'fri','sat','sun']
    i = 0
    
    while True:
        x = l[i]
        i = (i + 1) % 7
        yield x
        
d = Days()
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
print(next(d))
```

### Local vs Global Variables

- Global variables can be accessed from anywhere in the python file
- Local variable can only be accessed from within the function

```python
# Interactions of local and global variables

g = 10

def func():
    g = 20
    print(g)  # This will call the local variable instead of the global, global is unchanged
    
    
func()
print(g)
```

```python
# A function cannot modify a global variable normally, it can only access it
# In order for your function to modify the global variable, use the global keyword

g = 10

def func():
    global g # This will allow your function to modify the global variable
    g = 20 
    print(g)  
    
    
func()
print(g)
```
