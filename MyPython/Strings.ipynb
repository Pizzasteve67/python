### Challenge Display data in given format (25 letters)

Display Data in a given format (Data and Price total 25 characters of space)

```python
item = input('Input data of item')
price = input('Input price of item')

length = len(item) + len(price)

if length > 25:
    print("Too long")
else:
    dots = "." * (25-length)
    print(item + dots + price)
```

### Challenge Palindrome

- Check if string is a palindrome
- Convert a given string to palindrome

```python
s1 = input("enter your string")
invert = ""

for i in range(len(s1)-1,-1, -1):
    invert = invert + s1[i]
    
if invert == s1:
    print("Palindrome")
else:
    print("Not a Palindrome")
    
```

```python
s2 = input("enter your string")
invert2 = ""

for i in range(len(s2)-1,-1, -1):
    invert2 = invert2 + s2[i]

palindrome = s2 + invert2
print(palindrome)
```

### Challenge Check if two strings are Anagram

decimal is anagram to medical

```python
s1 = input("input first string")
s2 = input("input second string")

s1 = sorted(s1)
s2 = sorted(s2)

if len(s1) != len(s2):
    print("Not Anagram")
elif s1 == s2:
    print("Anagram")
else:
    print("Not Anagram")
```

```python
s1 = input("input first string")
s2 = input("input second string")

if len(s1) != len(s2):
    print("Not Anagram")
else:
    for i in s1:
        if i not in s2:
            print("Not Anagram")
            break
    else:
        print("Anagram")
```

### String Challenge

Rearrange lowercase then uppercase<br>

'AbcDEfgHy'<br>
lower case first then uppercase

```python
s = 'AbcDEfgHy'
lower = ''
upper = ''

for i in s:
    if i.isupper():
        upper = upper + i
    else:
        lower = lower + i
        
new = sorted(lower) + sorted(upper)
arranged = "".join(new)
arranged 
```
